```python

#Importing libraries we need
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


 
 
```

```python
#import and read data
irisData = pd.read_csv('iris.csv')
```

### Sepal length by Sepal width by species

```python

versiLen = irisData[irisData['species']=='versicolor'].sepal_length
versiWidth = irisData[irisData['species']=='versicolor'].sepal_width
plt.plot(versiWidth,versiLen,color='green',  linestyle='solid',linewidth=2, markersize=12)
plt.xlabel("sepal_width")
plt.ylabel("sepal_length")
plt.title('Versicolor sepal')
plt.show()
```

```python
setosaLen = irisData[irisData['species']=='setosa'].sepal_length
setosaWidth = irisData[irisData['species']=='setosa'].sepal_width

plt.plot(setosaWidth,setosaLen, linestyle='solid',
     linewidth=2, markersize=12)
plt.xlabel("sepal_width")
plt.ylabel("sepal_length")
plt.title('Setosa sepal')
plt.show()
```

```python
virgiLen = irisData[irisData['species']=='virginica'].sepal_length
virgiWidth = irisData[irisData['species']=='virgica'].sepal_width
plt.plot(versiWidth,virgiLen, color='orange',  linestyle='solid',
     linewidth=2, markersize=12)
plt.xlabel("sepal_width")
plt.ylabel("sepal_length")
plt.title('Virginica sepal')
plt.show()
```
