```python
from bs4 import BeautifulSoup
import requests
```

```python
#scrape data from sources
url = "https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&txtKeywords=python&txtLocation="
page = requests.get(url).text
```

```python
soup = BeautifulSoup(page,'lxml')
jobs = soup.find_all('li',class_ = "clearfix job-bx wht-shd-bx")
```

```python
for job in jobs:
    publishtime = job.find('span',class_ ="sim-posted").span.text
    if 'few' in  publishtime :
        company = job.find('h3',class_ ="joblist-comp-name").text
        skills = job.find('span',class_ ="srp-skills").text.replace(' ','')
        links = job.header.h2.a['href']
        if 'MORE JOBS'in company:
            company = company.replace('MORE JOBS','')
        print( f"\033[1m Company Name: \033[0m {(company.strip()).capitalize().upper()}")
        print(f"\033[1m Required Skills: \033[0m {skills.strip().capitalize()}") 
        print(f"\033[1m Publication Date: \033[0m {publishtime } ")
        print((f"\033[1m Link: \033[0m {links } \n"))
     
```
